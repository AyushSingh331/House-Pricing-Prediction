```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
print(df.isna().sum().to_string())
```

```python
df["HasGarage"]=df["GarageType"].isna().astype(int)
df["HasGarage"]=1-df["HasGarage"]
df["HasGarage"].value_counts()
```

```python
df["HasBsmt"]=df["BsmtQual"].isna().astype(int)
df["HasBsmt"]=1-df["HasBsmt"]
df["HasBsmt"].value_counts()
```

```python
df["LotFrontage"]=df["LotFrontage"].fillna(0)
df["Electrical"]=df["Electrical"].fillna("Mix")
```

```python
df["LotFrontage"].info();
```

```python
df["Electrical"].info()
```

```python
df=df.drop(["Alley","MasVnrType","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","GarageQual",
            "PoolQC","Fence","MiscFeature","FireplaceQu","GarageYrBlt","GarageType","GarageFinish","GarageCond"],axis=1)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

object_cols=df.select_dtypes(include=["object"]).columns
for col in object_cols:
    labelEncoder=LabelEncoder()
    df[col]=labelEncoder.fit_transform(df[col].astype(str)).astype("int64")

df.info()
```

## Fitting The Values to XGBoost

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import xgboost as xgb

x=df.drop("SalePrice",axis=1)
y=df["SalePrice"]

x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=42)

xgb_model=xgb.XGBRegressor()
xgb_model.fit(x_train,y_train)

xgb_model_pred=xgb_model.predict(x_val)
xgb_model_mae=mean_absolute_error(xgb_model_pred,y_val)

print("Mean Absolute Error(XGBoost):",xgb_model_mae)
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid={"n_estimators":[100,200,300],
            "max_depth":[3,5,7,9],
            "learning_rate":[0.01,0.1,0.2,0.3],
            "subsample":[0.7,0.9],
            "colsample_bytree":[0.5,0.7,0.9]}

grid_search=GridSearchCV(xgb.XGBRegressor(),param_grid=param_grid,cv=5,scoring="neg_mean_absolute_error")
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
xgb_model_grid=xgb.XGBRegressor(colsample_bytree=0.7,learning_rate=0.1,max_depth=3,
                                n_estimators=300,subsample=0.7,random_state=42)
xgb_model_grid.fit(x_train,y_train)

xgb_model_grid_pred=xgb_model_grid.predict(x_val)
xgb_model_grid_mae=mean_absolute_error(xgb_model_grid_pred,y_val)

print("Mean Absolute Error(XGBoost)",xgb_model_grid_mae)
```
